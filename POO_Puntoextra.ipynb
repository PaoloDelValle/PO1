```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1.9, -1.5, 1.5, -1.9, 0, 1.9, -1.5, 1.5, -1.9, 0, 1.4, -1.1, 1.1, -1.4, 0, 1.4, -1.1, 1.1, -1.4]
y = [3, -1, 0.5, 0.5, -1, -3, 1, -0.5, -0.5, 1, 2, -0.7, 0.35, 0.35, -0.7, -2, 0.7, -0.35, -0.35, 0.7]


plt.scatter(x, y, label= 'Estrella')

plt.xlabel('X')
plt.xlabel('Y')
plt.title('Rotar una estrella 88°')
plt.legend()

theta = np.radians(88)
matriz_rotacion = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
imagen_rotacion = np.dot(matriz_rotacion, np.array([x, y]))

print(f"La matriz de rotacion por 88° es: (imagen_rotacion)")

plt.scatter(imagen_rotacion[0], imagen_rotacion[1], label='Rotacion', marker='x')
plt.show()
```
